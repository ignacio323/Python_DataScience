```python
import matplotlib.pyplot as plt
import numpy as np

# Función para el costo anual de download music con Music A
def cost_music_A(x):
    return 10 + 1.25 * x

# Función para el costo anual de download music con Music B
def cost_music_B(x):
    return 14 + 0.75 * x

# Rango de valores para x (números enteros de 0 a 18)
x_values = np.arange(0, 19, 2)

# Calcula los valores de y para cada función
y_values_A = cost_music_A(x_values)
y_values_B = cost_music_B(x_values)

# Puntos seleccionados para graficar (primer y segundo punto)
point_A1 = (0, cost_music_A(0))
point_A2 = (10, cost_music_A(10))

point_B1 = (0, cost_music_B(0))
point_B2 = (10, cost_music_B(10))

# Grafica las funciones
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values_A, label='Music A')
plt.plot(x_values, y_values_B, label='Music B')

# Marca los puntos seleccionados
plt.scatter(*point_A1, color='red')
plt.scatter(*point_A2, color='red')

plt.scatter(*point_B1, color='blue')
plt.scatter(*point_B2, color='blue')

# Configuración del gráfico
plt.grid(True)
plt.xlabel('Número de descargas')
plt.ylabel('Costo anual (en dólares)')
plt.title('Costo anual de download music con Music A y Music B')
plt.xticks(np.arange(0, 19, 2))
plt.yticks(np.arange(0, 20, 2))
plt.legend()

# Muestra el gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Function to calculate the cost for Music A
def cost_music_a(downloads):
    return 10 + 1.25 * downloads

# Function to calculate the cost for Music B
def cost_music_b(downloads):
    return 14 + 0.75 * downloads

# Generate x values for the graph
x_values = np.arange(0, 20, 2)

# Calculate y values for both Music A and Music B
y_values_a = [cost_music_a(x) for x in x_values]
y_values_b = [cost_music_b(x) for x in x_values]

# Plot the graph
plt.figure(figsize=(8, 8))
plt.plot(x_values, y_values_a, label='Music A')
plt.plot(x_values, y_values_b, label='Music B')

# Set axis labels and title
plt.xlabel('Number of Downloads')
plt.ylabel('Annual Cost')
plt.title('Annual Cost of Downloading Music')

# Set axis ticks and grid
plt.xticks(np.arange(0, 20, 2))
plt.yticks(np.arange(0, 20, 2))
plt.grid(True)

# Show legend
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Function for Company A's total cost
def company_a_cost(x):
    return 40 + 0.80 * x

# Function for Company B's total cost
def company_b_cost(x):
    return 27 + 1.20 * x

# Choose two points for each company
points_x = [0, 10]
points_a_y = [company_a_cost(x) for x in points_x]
points_b_y = [company_b_cost(x) for x in points_x]

# Plot the points
plt.plot(points_x, points_a_y, marker='o', label='Company A')
plt.plot(points_x, points_b_y, marker='o', label='Company B')

# Add labels and title
plt.xlabel('Number of Engraved Letters')
plt.ylabel('Total Cost')
plt.title('Total Cost of Engraved Plaque from Company A and Company B')

# Add legend
plt.legend()

# Show the plot
plt.show()
```
