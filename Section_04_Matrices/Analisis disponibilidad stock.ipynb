```python
!pip install pandas
import pandas as pd

# Ruta al archivo Excel local
ruta_archivo = 'C:/Users/User1/Desktop/Python Course/supply_chain_data.csv'
# Cargar el archivo Excel en un DataFrame de pandas
df = pd.read_csv(ruta_archivo)
df.head() # Visualizar las primeras filas del DataFrame

# Anlisis de Disponibilidad y Stock
analisis_disponibilidad_stock = df[['SKU', 'Availability', 'Stock levels', 'Number of products sold']]

descripcion = analisis_disponibilidad_stock.describe() # Estadsticas descriptivas

# Visualizacin de tendencias
import matplotlib.pyplot as plt
import seaborn as sns

# Grfico de disponibilidad y stock
plt.figure(figsize=(12, 6))
sns.lineplot(data=analisis_disponibilidad_stock, x='SKU', y='Availability', label='Availability')
sns.lineplot(data=analisis_disponibilidad_stock, x='SKU', y='Stock levels', label='Stock levels')
plt.title('Disponibilidad y Stock por SKU')
plt.xlabel('SKU')
plt.ylabel('Cantidad')
plt.legend()
plt.show()
```
