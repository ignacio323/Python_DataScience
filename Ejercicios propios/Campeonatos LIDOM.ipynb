```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Crear un Array con los valores de la lista anterior.
licey_champs     = [1951, 1953, 1959, 1964, 1970, 1971, 1973, 1974, 1977, 1980, 1983, 1984, 1985, 1991, 1994, 1999, 2002, 2004, 2006, 2009, 2014, 2017, 2023]
aguilas_champs   = [1952, 1965, 1967, 1972, 1975, 1976, 1978, 1979, 1986, 1987, 1993, 1996, 1997, 1998, 2000, 2001, 2003, 2005, 2007, 2008, 2018, 2021]
escogido_champs  = [1956, 1957, 1958, 1960, 1961, 1969, 1981, 1982, 1988, 1989, 1990, 1992, 2010, 2012, 2013, 2016]
estrellas_champs = [1954, 1968, 2019]
toros_champs     = [1995, 2011, 2020]
gigantes_champs  = [2015, 2022]
print("   Campeonatos ganados por el Licey: " , len(licey_champs))
print("Campeonatos ganados por las Aguilas: " , len(aguilas_champs))
print(" Campeonatos ganados por los Leones: " , len(escogido_champs))

# 2. Dictionarys 
teams_champs = {
    'Licey'    : licey_champs,
    'Aguilas'  : aguilas_champs,
    'Escogido' : escogido_champs,
    'Estrellas': estrellas_champs,
    'Toros'    : toros_champs,
    'Gigantes' : gigantes_champs
}

team_names = ['Licey', 'Aguilas', 'Escogido', 'Estrellas', 'Toros', 'Gigantes']
team_colors = ['blue', 'yellow', 'red', 'green' , 'orange', 'purple']

# 3. Crear un Array con los valores de todos los a√±os, desde el 1951 hasta el 2024.
all_years = np.arange(1951, 2025)

# 4. Imprimir los equipos que han ganado campeonatos en a√±os consecutivos.
consecutive_winners = []

for team, champs in teams_champs.items():
    for year in champs:
        if year + 1 in champs:
            consecutive_winners.append((team, year, year + 1))

print("Equipos que han ganado campeonatos en a√±os consecutivos:")
for item in consecutive_winners:
    print(f"{item[0]}: {item[1]} - {item[2]}")

# 4. Generar todos los gr√°ficos posibles que estas estad√≠sticas pueden ofrecer.

# Gr√°fico de barras para mostrar la cantidad de campeonatos ganados por cada equipo.
team_names = list(teams_champs.keys())
champs_count = [len(champs) for champs in teams_champs.values()]

plt.figure(figsize=(10, 6))
plt.bar(team_names, champs_count, color = team_colors)
plt.title('Campeonatos Ganados por Equipo')
plt.xlabel('Equipos')
plt.ylabel('Campeonatos Ganados')
# plt.show()

# Gr√°fico de l√≠neas para mostrar la distribuci√≥n de campeonatos a lo largo de los a√±os.
plt.figure(figsize=(15, 8))
for i, (team, champs) in enumerate(teams_champs.items()):
    plt.plot(champs, [i] * len(champs), marker='o', linestyle='', label=team, color=team_colors[i])
    
plt.title('Distribuci√≥n de Campeonatos a lo largo de los A√±os')
plt.xlabel('A√±os')
plt.ylabel('Equipos')
plt.legend()
plt.show()
```
