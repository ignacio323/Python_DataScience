```python
import pandas as pd

# Cargar el conjunto de datos desde un archivo CSV local
nba_data = pd.read_csv('C:\\Users\\User1\\Desktop\\Python Course\\Ejercicios propios\\NBA\\players_stats.csv'
# Solicitar al usuario que ingrese los nombres de los dos equipos
equipo1 = input("Ingresa el nombre del primer equipo: ")
equipo2 = input("Ingresa el nombre del segundo equipo: ")

# Filtrar el dataframe para obtener las estadísticas de los dos equipos
stats_equipo1 = nba_data[nba_data['TEAM'].str.contains(equipo1)].mean()
stats_equipo2 = nba_data[nba_data['TEAM'].str.contains(equipo2)].mean()

# Calcular la proyección para un juego de 48 minutos
minutos_por_juego = 48
proyeccion_equipo1 = stats_equipo1 * minutos_por_juego
proyeccion_equipo2 = stats_equipo2 * minutos_por_juego

# Imprimir los resultados proyectados
print(f"Proyección para {equipo1} en un juego de 48 minutos:")
print(proyeccion_equipo1)

print(f"\nProyección para {equipo2} en un juego de 48 minutos:")
print(proyeccion_equipo2)

                       
```
